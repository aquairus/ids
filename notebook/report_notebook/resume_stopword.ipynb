```python
frequent_words=dict()
for i in range(num_clusters):
    for j in range(i+1,num_clusters):
        class_1=km.cluster_centers_[i]>0.001
        class_2=km.cluster_centers_[j]>0.001
        intersection=class_1 & class_2

        for idx in intersection.nonzero()[0]:
            word=vectorizer.get_feature_names()[idx]
            if word in frequent_words:
                frequent_words[word]+=1
            else:
                frequent_words[word]=1
                
```
